#### First look of the data:
- 8 columns, all in object dtype -> Need to change some of them to suitable dtype
- There are null values -> Need to check if the values are in the correct NA form

#### Changing column names

```python
df.rename(columns={'Price Per Unit': 'Unit Price', 'Transaction Date': 'Date', 'Payment Method': 'Payment Mode'}, inplace=True)
df.columns
```

```python
# Transaction ID does not seem important for this current project, it is safe to remove this column
# This is optional and heavily depends on the project needs.

df.drop(['Transaction ID'],axis=1, inplace=True)
df.columns # To check that 'Transaction ID' is removed from df
```

#### Change null values to pandas friendly null values
- This is done so that it is easier to manipulate data later (such as changing data type, doing arithmetic operation)

```python
df['Date'].unique()

# ERROR and UNKNOWN seems to be the wrong data input for null values
```

```python
# Identifying non-values input (This is repeated for each column)
# The basic code would be:
# df['Location'].unique()
# Here, we use for loop, instead of repeating the same code for each column

cols= ['Item', 'Quantity', 'Unit Price', 'Total Spent','Payment Mode', 'Location']

for x in df[cols]:
    print('Unique values in {} columns: {}\n'.format(x,df[x].unique()))
    
# It seems that  UNKNOWN , ERROR are also used as null values in other columns
```

#### Change data type
- float dtype: Quantity, Price Per Unit,Total Spent
- datetime dtype: Date
- object dtype: others

```python
# To change dtype, ensure the non-values are in the correct python Nan/None first

dict_replace = {'UNKNOWN':np.nan, 'ERROR':np.nan}

for col in df.columns:
    df[col] = df[col].replace(dict_replace)
```

```python
df['Quantity'] = df['Quantity'].astype('float') 
df['Unit Price'] = df['Unit Price'].astype('float')
df['Total Spent'] = df['Total Spent'].astype('float')
df['Date'] = pd.to_datetime(df['Date'])

# Changing Quantity column to float as np.nan type is float. It is easier to manipulate data later on.
```

```python
# Let's check the dtype

df.info()
```

```python
# Removing null values in Date and Item columns 
# Null values in Item make up 0.09% of the whole data, dropping them might be safe and not skewing the data.

df.dropna(subset=['Date','Item'], inplace=True)
```

```python
# For the sake of having more data for visualizations, 
# let's fill the categorical column nans with forward or backward fill (either one is fine)

df['Payment Mode'].ffill(inplace=True)
df['Location'].bfill(inplace=True)
```

```python
df.info()

# So, we have a total of 8613 rows of data now, all categorical and date columns are filled. 
# Lets' fill in the numerical columns
```

#### Dropping rows with at least 2 data incomplete in the numerical columns
- For example:
- if Quantity AND Unit Price are Nan, the row is dropped

```python
# Here, we are subsetting the rows with Null values in at least 2 numerical columns

qty_price = df[df['Quantity'].isna() & df['Unit Price'].isna()]
qty_spent = df[df['Quantity'].isna() & df['Total Spent'].isna()]
price_spent = df[df['Unit Price'].isna() & df['Total Spent'].isna()]
```

```python
# Here, we are removing the above rows

df = df.drop(qty_price.index)
df = df.drop(qty_spent.index)
df = df.drop(price_spent.index)
```

```python
df.shape

# 8613 rows of data has been reduced to 8564 rows
```

## Time to impute null the values in numerical columns

- Total spent = qty * unit price
- qty = total spent / unit price
- unit price = total spent/qty


```python
df['Total Spent'] = np.where(df['Total Spent'].isnull(), df['Quantity']*df['Unit Price'], df['Total Spent'])
df['Quantity'] = np.where(df['Quantity'].isnull(), df['Total Spent']/df['Unit Price'], df['Quantity'])
df['Unit Price'] = np.where(df['Unit Price'].isnull(), df['Total Spent']/df['Quantity'], df['Unit Price'])
```

### What food is most popular (can divide quarterly)

```python
print('This data starts from {} to {}.'.format(df['Date'].min(), df['Date'].max()))
```

```python
q1 = df[df['Date'].dt.month.isin([1,2,3])]
q2 = df[df['Date'].dt.month.isin([4,5,6])]
q3 = df[df['Date'].dt.month.isin([7,8,9])]
q4 = df[df['Date'].dt.month.isin([10,11,12])]
```

```python
def popular_food(data,name):
    '''
    Plotting highest bought food item based on each quarter.
    '''
    fig, ax = plt.subplots(figsize = (5,3))
    sns.boxplot(x='Item',y='Total Spent', data=data)
    ax.set_title(label=name, fontsize=25, weight='bold')
    ax.set_xlabel(xlabel='Food Item', fontsize=12)
    ax.tick_params(axis='x', labelrotation=30)
    plt.show()
    
#popular_food(q1,'Quarter 1')
```

```python
# Set a list of tuples of function parameter
q_list = [(q1,'Quarter 1'),(q2,'Quarter 2'),(q3,'Quarter 3'),(q4,'Quarter 4')]

for x,y in q_list:
    popular_food(x,y)
```

#### Let's create a new column to label our data quarterly (from q1 to q4)

```python
def quarterly(dataframe: pd.DataFrame):
    """
    Classify data for each quarter of the year
    """
    
    if dataframe['Date'].month in [1,2,3]:
        return 'Q1'
    elif dataframe['Date'].month in [4,5,6]:
        return 'Q2'
    elif dataframe['Date'].month in [7,8,9]:
        return 'Q3'
    elif dataframe['Date'].month in [10,11,12]:
        return 'Q4'
```

```python
df['Quarter'] = df.apply(quarterly, axis=1)
```

```python
ax = sns.relplot(x='Quarter', y='Total Spent', data=df, kind='line', errorbar=None).set(title='Mean of Total Spent vs Yearly Quarter')
plt.show()
```
